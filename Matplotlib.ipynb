```python
#  3차원 


import numpy as np

np.random.seed(0) # 난수를 일정한 상태로 고정시킨다

n = 50

x = np.random.rand(n)
y = np.random.rand(n)
z = np.random.rand(n)

area = (30*np.random.rand(n)) ** 2
colors = np.random.rand(n)

fig = plt.figure(figsize=(6,6))  # 액자
ax = fig.add_subplot(111, projection='3d') # 사진
ax.scatter(x,y,z, s= area, c=colors, alpha = 0.4, cmap='Set1') # 액자+사진
# plt.colorbar()
plt.show # 출력
```
